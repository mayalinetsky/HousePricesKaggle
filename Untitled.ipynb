```python
import pandas as pd 
from matplotlib import pyplot as plt 
from sklearn.model_selection import train_test_split

train_df = pd.read_csv('train.csv').select_dtypes(include='number')
train, validation = train_test_split(train_df, test_size=0.2, random_state=0)

for column in train_df.columns: 
    # print(column)
    fig, ax = plt.subplots(1,3,figsize=(12,3)) 
    ax[0].hist(train_df[column], density=True)
    ax[0].set_title(f'{column} (whole)')
    ax[1].hist(train[column], density=True)
    ax[1].set_title(f'{column} (subset_train)')
    ax[2].hist(validation[column], density=True)
    ax[2].set_title(f'{column} (subset_validation)')  
    plt.show()
    
```
